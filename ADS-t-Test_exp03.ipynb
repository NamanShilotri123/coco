```python
PrePostScore=pd.read_excel(r'/content/Pre_Post_Score.xlsx')
PrePostScore
```

```python
Two_Sample_Result=stats.ttest_rel(PrePostScore["Pre_Score"],PrePostScore["Post_Score"])
Two_Sample_Result
```

Conclusion:Two sample paired t-test has been done on the prescore-post score dataset.Since the difference occurs between the calculated value and critical value, therefore there exists a difference between the score before learning the module and after learning the module.
